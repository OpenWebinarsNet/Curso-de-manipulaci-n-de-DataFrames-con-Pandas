```python
# Join and mergering

import numpy as np
import pandas as pd
```

```python
productos = pd.DataFrame({'Codigo_producto': ['K0', 'K1', 'K2', 'K3', 'K5'],
                     'Ventas': [600, 450, 850, 600, 400],
                     'Devoluciones': [50, 30, 40, 25, 30]})
   
productos_aa = pd.DataFrame({'Codigo_producto': ['K0', 'K1', 'K2', 'K4'],
                          'Ventas AA': [500, 800, 600, 550],
                          'Devoluciones AA': [10, 22, 18, 31]})  
```

```python
pd.merge(productos, productos_aa, how='inner', on='Codigo_producto', indicator=True)
```

```python
# Otra forma de hacer lo mismo pero trabajando con la columna _merge que se crea a partir de indicator
df_merge = pd.merge(productos, productos_aa, how='inner', on='Codigo_producto', indicator=True)
df_merge = df_merge[df_merge['_merge'] == 'both'].drop(['_merge'], axis=1)
df_merge
```

```python
pd.merge(productos, productos_aa, how='left', on='Codigo_producto', indicator=True)
```

```python
# Si quisera quedarme únicamente con los registros que solo se encuentran en el df de la derecha
df_merge = pd.merge(productos, productos_aa, how='left', on='Codigo_producto', indicator=True)
df_merge = df_merge[df_merge['_merge'] == 'left_only'].drop(['_merge'], axis=1)
df_merge
```

```python
pd.merge(productos, productos_aa, how='right', on='Codigo_producto', indicator=True)
```

```python
# Si quisera quedarme únicamente con los registros que solo se encuentran en el df de la derecha
df_merge = pd.merge(productos, productos_aa, how='right', on='Codigo_producto', indicator=True)
df_merge = df_merge[df_merge['_merge'] == 'right_only'].drop(['_merge'], axis=1)
df_merge
```

```python
pd.merge(productos, productos_aa, how='outer', on='Codigo_producto', indicator=True)
```

```python
# Un ejemplo más complicado
# Hacer join por dos campos
```

```python
productos = pd.DataFrame({'Codigo_producto': ['K0', 'K0', 'K1', 'K2'],
                     'Codigo_subproducto': ['K0', 'K1', 'K0', 'K1'],
                        'A': [500, 800, 600, 550],
                        'B': [10, 22, 18, 31]})
    
productos_aa = pd.DataFrame({'Codigo_producto': ['K0', 'K1', 'K1', 'K2'],
                      'Codigo_subproducto': ['K0', 'K0', 'K0', 'K0'],
                       'C': [130, 500, 500, 800],
                       'D': [10, 22, 18, 31]})
```

```python
# Por defecto hace un inner 
# how = 'inner'
pd.merge(productos, productos_aa, 
         on=['Codigo_producto', 'Codigo_subproducto'], indicator=True)
```

```python
# Por defecto hace un inner 
# how = 'right'
pd.merge(productos, productos_aa,
         how='right', 
         on=['Codigo_producto', 'Codigo_subproducto'], indicator=True)
```

```python
# how = 'left'
pd.merge(productos, productos_aa,
         how='left', 
         on=['Codigo_producto', 'Codigo_subproducto'], indicator=True)
```

```python
df_empresa1 = pd.DataFrame({'A': [10, 22, 18],
                     'B': [25, 14, 23]},
                      index=['K0', 'K1', 'K2']) 

df_empresa2 = pd.DataFrame({'C': [27, 28, 25],
                    'D': [21, 18, 26]},
                      index=['K0', 'K2', 'K3'])
```

```python
# Por defecto hará un left join
df_empresa1.join(df_empresa2)
```

```python
# Outer join
df_empresa1.join(df_empresa2, how='outer')
```

```python
df_empresa1.join(df_empresa2, how='right')
```
